```python
import pandas as pd

df1 = pd.read_csv("/content/merged_core_it_courses_fixed.csv")
df2 = pd.read_csv("/content/courses.csv")

courses1 = pd.DataFrame({
    "course_title": df1["course_title"],
    "provider": df1["provider"],
    "level": df1["level"],
    "rating": df1.get("rating"),
    "reviews": df1.get("reviews"),
    "url": df1["url"],
    "description": df1["description"],
    "skills": df1["skills"],
    "track": df1.get("it_label"),
    "prerequisites": "",
    "roadmaps": "",
    "source_dataset": df1.get("source_dataset", "merged_core_it_courses"),
    "quality_weight": df1.get("quality_weight", 1.0)
})

courses2 = pd.DataFrame({
    "course_title": df2["name"],
    "provider": df2["provider"],
    "level": df2["level"],
    "rating": None,
    "reviews": None,
    "url": df2["link"],
    "description": df2["summary"],
    "skills": (
        df2["track"].fillna("") + " " +
        df2["prerequisites"].fillna("")
    ),
    "track": df2["track"],
    "prerequisites": df2["prerequisites"],
    "roadmaps": df2["roadmaps"],
    "source_dataset": "courses_csv",
    "quality_weight": 0.82
})

all_courses = pd.concat([courses1, courses2], ignore_index=True)

all_courses["course_title"] = all_courses["course_title"].fillna("").str.strip()
all_courses["description"] = all_courses["description"].fillna("")
all_courses["skills"] = all_courses["skills"].fillna("")
all_courses["provider"] = all_courses["provider"].fillna("Unknown")
all_courses["level"] = all_courses["level"].fillna("Unknown")

all_courses = all_courses[all_courses["course_title"] != ""]

all_courses = all_courses.drop_duplicates(
    subset=["course_title", "provider"],
    keep="first"
)

all_courses.to_csv("/content/all_courses_combined_pathfinder.csv", index=False)

print("Final shape:", all_courses.shape)
all_courses.head()
```
